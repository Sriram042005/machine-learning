```python
import numpy as np
import matplotlib.pyplot as mtp
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Load the diabetes dataset
diabetes = datasets.load_diabetes()

# Print the feature names
print("Feature names:", diabetes['feature_names'])

# Print the first five rows of the data
print("First five rows of data:\n", diabetes['data'][:5])

# Define the features (independent variables) and target (dependent variable)
x = diabetes.data[:, np.newaxis, 2]  # Use a single feature for simplicity (e.g., 2nd feature)
y = diabetes.target  # The target variable is 'target'

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state=0)

# Initialize the linear regression model
regressor = LinearRegression()

# Fit the model on the training data
regressor.fit(x_train, y_train)

# Make predictions on the test and train data
y_pred_test = regressor.predict(x_test)
y_pred_train = regressor.predict(x_train)

# Visualize the training data and the linear regression line
mtp.scatter(x_train, y_train, color="green")  # Scatter plot for training data
mtp.plot(x_train, y_pred_train, color="red")  # Line plot for the regression line
mtp.title('Linear Regression on Diabetes Dataset')
mtp.xlabel('Feature: BMI')
mtp.ylabel('Target: Disease Progression')
mtp.show()

# Optional: Print the coefficients and the intercept
print("Coefficient (slope):", regressor.coef_)
print("Intercept:", regressor.intercept_)
```
